```python
import pandas as pd
```

```python
df= pd.read_csv(r"Data\World_ranking.csv")
```

```python
df_definitivo= df[["FIFA_Rank", "Country"]]
df_definitivo=df_definitivo.rename(
    columns={
        "FIFA_Rank" :"Equipo",
        "Country": "Rating"
    }
)
```

```python
df2= pd.read_csv(r"Data\wikipedia-iso-country-codes.csv")
df_tabla = df2[["English short name lower case","Alpha-2 code"]]
df_tabla= df_tabla.rename(
    columns={
        "English short name lower case": "Nombre",
        "Alpha-2 code" : "Codigo"
    }
)
df_tabla
```

```python
df_final= pd.merge(
    df_definitivo,
    df_tabla,
    left_on="Equipo",
    right_on="Codigo",
    how="left"
)

df_final
```

```python
lista= ["Nombre","Rating"]
df_final= df_final[lista]
df_final= df_final.rename(
    columns={
        "Nombre":"Equipo"
    }
)
```

```python
df_final.to_csv(r"output\fifa_elo.csv", index=False)
```
