# CSV Data Explorer Notebook

This notebook automatically scans the **Data** directory for CSV files, loads each into a pandas DataFrame, and displays a quick summary including the first few rows, column data types, and basic statistics.

**Directory:** `c:/Users/fcoer/Desktop/exp world cup/Data`

```python
import os
import glob
import pandas as pd
from IPython.display import display, HTML

# Path to the data directory
data_dir = r'c:/Users/fcoer/Desktop/exp world cup/Data'

# Find all CSV files in the directory
csv_files = glob.glob(os.path.join(data_dir, '*.csv'))
print(f'Found {len(csv_files)} CSV files:')
for f in csv_files:
    print(' -', os.path.basename(f))
```

```python
def summarize_csv(file_path):
    df = pd.read_csv(file_path)
    print(f'\n--- Summary for {os.path.basename(file_path)} ---')
    display(HTML(df.head().to_html(index=False)))
    print('Shape:', df.shape)
    print('Columns and dtypes:')
    print(df.dtypes)
    print('Numeric columns summary:')
    numeric_cols = df.select_dtypes(include='number').columns
    if len(numeric_cols):
        display(df[numeric_cols].describe().to_html())
    else:
        print('No numeric columns.')
    print('\n')
    return df
```

```python
dataframes = {}
tournaments = set()
for file in csv_files:
    df = summarize_csv(file)
    dataframes[os.path.basename(file)] = df
    if 'tournament' in df.columns:
        tournaments.update(df['tournament'].dropna().unique())

print('Unique tournaments found:')
for t in sorted(tournaments):
    print(' -', t)
```

```python
# Crear un DataFrame con solo los datos del torneo "Fifa world cup qualification" entre 2022 y 2026
results_df = dataframes.get('results.csv')
if results_df is None:
    results_path = os.path.join(data_dir, 'results.csv')
    results_df = pd.read_csv(results_path)

results_df['date'] = pd.to_datetime(results_df['date'], errors='coerce')
fifa_wcq_df = results_df[
    (results_df['tournament'] == 'FIFA World Cup qualification') &
    (results_df['date'].dt.year >= 2022) &
    (results_df['date'].dt.year <= 2026)
]
print(f'Found {len(fifa_wcq_df)} rows for Fifa World Cup Qualification 2022-2026')
display(HTML(fifa_wcq_df.head().to_html(index=False)))
```
