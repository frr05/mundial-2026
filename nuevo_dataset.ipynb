```python
import pandas as pd

# Cargar dataset
matches =  pd.read_csv(r"Data\results.csv")

# Convertir fecha
matches['date'] = pd.to_datetime(matches['date'])

# Ver información original
print("DATASET ORIGINAL")
print(f"Partidos: {len(matches)}")
print(f"Fechas: {matches['date'].min()} a {matches['date'].max()}")

# ---------------------------------------------------
# FILTRAR SOLO PARTIDOS DESPUÉS DEL MUNDIAL 2022
# ---------------------------------------------------

fecha_corte = pd.Timestamp('2018-01-01')

matches = matches[matches['date'] >= fecha_corte].copy()

# Reiniciar índice
matches.reset_index(drop=True, inplace=True)

# ---------------------------------------------------
# LIMPIEZA BÁSICA
# ---------------------------------------------------

# Eliminar duplicados
matches.drop_duplicates(inplace=True)

# Ordenar por fecha
matches.sort_values('date', inplace=True)

# Revisar valores nulos
print("\nVALORES NULOS")
print(matches.isnull().sum())

# Información final
print("\nDATASET LIMPIO")
print(f"Partidos: {len(matches)}")
print(f"Fechas: {matches['date'].min()} a {matches['date'].max()}")

# Ver columnas
print("\nCOLUMNAS:")
print(matches.columns.tolist())
```

```python
matches.to_csv("Data/Matches.csv", index=False)
```
