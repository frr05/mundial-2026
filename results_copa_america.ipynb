# Results – Copa América Filter

This notebook loads the **results.csv** dataset and filters rows where the `tournament` column equals **Copa América**. It then displays the filtered data and a quick summary.

```python
import pandas as pd
from IPython.display import display, HTML

# Path to the results CSV file
results_path = r'c:/Users/fcoer/Desktop/exp world cup/Data/results.csv'

# Load the dataset
df = pd.read_csv(results_path)
print(f'Loaded {len(df)} rows from results.csv')
```

```python
copa_america_df = df[df['tournament'] == 'Copa América']
print(f'Found {len(copa_america_df)} rows for Copa América')
display(HTML(copa_america_df.head().to_html(index=False)))
```

```python
# Convertir la columna date a datetime y filtrar los años 2015 y 2016 dentro de Copa América
copa_america_df['date'] = pd.to_datetime(copa_america_df['date'], errors='coerce')
copa_america_2015_2016_df = copa_america_df[copa_america_df['date'].dt.year.isin([2015, 2016])]
print(f'Found {len(copa_america_2015_2016_df)} rows for Copa América 2015-2016')
display(HTML(copa_america_2015_2016_df.head().to_html(index=False)))
```
